```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile
from urllib.request import urlretrieve

from IPython.display import Image

%matplotlib inline
```

```python
swallow_gray = cv2.imread('swallow.jpg', 0)
plt.imshow(swallow_gray, cmap="gray")
print(swallow_gray)
```

```python
print(swallow_gray[0,0])
print(swallow_gray[150,550])
```

```python
swallow_gray_copy = swallow_gray.copy()
swallow_gray_copy[100:350, 100:350] = 75

plt.imshow(swallow_gray_copy, cmap="gray")
print(swallow_gray_copy)
```

```python
swallow_bgr = cv2.imread('swallow.jpg', cv2.IMREAD_COLOR)
swallow_rgb = swallow_bgr[:, :,  ::-1]
plt.imshow(swallow_rgb)
```

```python
swallow_cropped = swallow_rgb[100:350, 400:650]
plt.imshow(swallow_cropped)
```

```python
swallow_resized_2x = cv2.resize(swallow_cropped, None, fx=2, fy=2)
plt.imshow(swallow_resized_2x)
```

```python
desired_width = 500
desired_height = 700
dim = (desired_width, desired_height)

swallow_resized_desired = cv2.resize(swallow_cropped, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(swallow_resized_desired)
```

```python
desired_width = 500
aspect_ration = desired_width / swallow_cropped.shape[1]
desired_height = int(swallow_cropped.shape[0] * aspect_ration)
dim = (desired_width, desired_height)

swallow_resized_desired_fixed = cv2.resize(swallow_cropped, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(swallow_resized_desired_fixed)
```

```python
cv2.imwrite('swallow_cropped_resized.png', swallow_resized_desired_fixed)
Image(filename='swallow_cropped_resized.png')
```

```python
swallow_flipped_h = cv2.flip(swallow_rgb, 1)
swallow_flipped_v = cv2.flip(swallow_rgb, 0)
swallow_flipped_b = cv2.flip(swallow_rgb, -1)

plt.figure(figsize=(18, 5))
plt.subplot(141);plt.imshow(swallow_flipped_h);plt.title("Horizontal Flip")
plt.subplot(142);plt.imshow(swallow_flipped_v);plt.title("Vertical Flip")
plt.subplot(143);plt.imshow(swallow_flipped_b);plt.title("Both Flipped")
plt.subplot(144);plt.imshow(swallow_rgb);plt.title("Original")
```
